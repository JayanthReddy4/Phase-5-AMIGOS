```python
#We will be using content based recommendation system
import pandas as pd
movies_df = pd.read_csv('/content/Full_data_feature _only.csv', na_values='missing_values')
ratings_df = pd.read_csv('/content/Full_data_ratings.csv', na_values='missing_values')
ratings_df['rating'] = ratings_df['rating'].astype(int)
print('Movies_df Shape:',movies_df.shape)
movies_df.head()
movies_df['Features'] = movies_df.Features.str.split(',')
movies_df.head()
ratings_df.head()
```

```python
# First let's make a copy of the movies_df.
movies_with_genres = movies_df.copy(deep=True)
x = []
for index, row in movies_df.iterrows():
    x.append(index)
    for genre in row['Features']:
        movies_with_genres.at[index, genre] = 1
        
# Confirm that every row has been iterated and acted upon.
print(len(x) == len(movies_df))
movies_with_genres.head(3)
```

```python
# so on a scale of 0 to 5, with 0 min and 5 max, see Lawrence's Store ratings below.
import pandas as pd
Lawrence_movie_ratings = [
            {'name':'AT&T Store', 'rating':4.9},
            {'name':'ALDO', 'rating':4.9},
         ] 
Lawrence_movie_ratings = pd.DataFrame(Lawrence_movie_ratings)
Lawrence_movie_ratings
```

```python
Lawrence_movie_ratings['rating'] = Lawrence_movie_ratings['rating'].astype(int)
Lawrence_movie_ratings
```

```python
# Extracting storeId Ids from movies_df and updating lawrence_movie_ratings with movie Ids.
Lawrence_movie_Id = movies_df[movies_df['name'].isin(Lawrence_movie_ratings['name'])]
Lawrence_movie_ratings = pd.merge(Lawrence_movie_Id, Lawrence_movie_ratings)
# Display the merged and updated data frame.
Lawrence_movie_ratings
```

```python
Lawrence_movie_ratings = Lawrence_movie_ratings.drop(['Features'], 1)
# Final profile for Lawrence
Lawrence_movie_ratings
```

```python
Lawrence_genres_df = movies_with_genres[movies_with_genres.store_Id.isin(Lawrence_movie_ratings.store_Id)]
Lawrence_genres_df
```

```python
# First, let's reset index to default and drop the existing index.
Lawrence_genres_df.reset_index(drop=True, inplace=True)
# Next, let's drop redundant columns
Lawrence_genres_df.drop(['store_Id','name','Features','vicinity'], axis=1, inplace=True)
# Let's view changes
Lawrence_genres_df
```

```python
# let's confirm the shapes of our data frames to guide us as we do matrix multiplication.
print('Shape of Lawrence_store_ratings is:',Lawrence_movie_ratings.shape)
print('Shape of Lawrence_genres_df is:',Lawrence_genres_df.shape)
#>>Shape of Lawrence_movie_ratings is: (9, 3) Shape of Lawrence_genres_df is: (9, 20)
```

```python
# Let's find the dot product of transpose of Lawrence_genres_df by Lawrence rating column.
Lawrence_profile = Lawrence_genres_df.T.dot(Lawrence_movie_ratings.rating)
# Let's see the result
Lawrence_profile
```

```python
# let's set the index to the storeId.
movies_with_genres = movies_with_genres.set_index(movies_with_genres.store_Id)
# let's view the head.
movies_with_genres.head()
```

```python
# Deleting four unnecessary columns.
movies_with_genres.drop(['store_Id','name','Features','vicinity'], axis=1, inplace=True)
# Viewing changes.
movies_with_genres.head()
```

```python
# Multiply the Features by the weights and then take the weighted average.
recommendation_table_df = (movies_with_genres.dot(Lawrence_profile)) / Lawrence_profile.sum()
```

```python
recommendation_table_df
```

```python
# first we make a copy of the original movies_df
copy = movies_df.copy(deep=True)
# Then we set its index to storeid
copy = copy.set_index('store_Id', drop=True)
# Next we enlist the top 20 recommended storeds we defined above
top_20_index = recommendation_table_df.index[:20].tolist()
# finally we slice these indices from the copied store df and save in a variable
recommended_movies = copy.loc[top_20_index, :]
# Now we can display the top 20 store in descending order of preference
recommended_movies
```
